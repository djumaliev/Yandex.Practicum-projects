```python
import pandas as pd
import nltk

from sklearn.linear_model import LogisticRegression
from nltk.corpus import stopwords as nltk_stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv('/Users/daniyardjumaliev/Jupyter/Projects/text_sentiment/tweets_lemm_train.csv')
test = pd.read_csv('/Users/daniyardjumaliev/Jupyter/Projects/text_sentiment/tweets_lemm_test.csv')

display(train)
display(test)
```

```python
corpus_train = train['lemm_text'].values.astype('U')
corpus_test = test['lemm_text'].values.astype('U')

display(corpus_train)
display(corpus_test)
```

```python
#nltk.download('stopwords')
#stopwords = set(nltk_stopwords.words('russian'))

count_tf_idf = TfidfVectorizer()
X_train = count_tf_idf.fit_transform(corpus_train)
y_train = train['positive']
X_test = count_tf_idf.transform(corpus_test)

display(X_train)
display(y_train)
display(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)

test['positive'] = model.predict(X_test)

test.to_csv('/Users/daniyardjumaliev/Jupyter/Projects/text_sentiment/predictions', index=False)

display(test)

accuracy = accuracy_score(y_train, model.predict(X_train))
print("Accuracy:", accuracy)
```
